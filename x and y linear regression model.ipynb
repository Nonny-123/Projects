```python
df['x'] = df['x'].astype(float)
```

```python
df['y'] = df['x'].astype(float)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('y', axis=1)
```

```python
y = df.drop('x', axis=1)
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,y, alpha=0.3)
plt.title("X and Y Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.ylim(0, 40)
plt.xlim(0, 10)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=3)
plt.title("X and Y Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.ylim(0, 40)
plt.xlim(0, 10)
plt.show()
```
