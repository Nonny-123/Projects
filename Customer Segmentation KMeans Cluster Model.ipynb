```python
for col in df.columns:
    temp = df[col].isnull().sum()
    if temp > 0:
        print(f"Column {col} contains {temp} null values")
```

```python
df = df.dropna()
```

```python
df.nunique()
```

```python
df.head()
```

```python
df[['Dt_day', 'Dt_month', 'Dt_year']] = df['Dt_Customer'].str.split("-",  expand=True)
```

```python
df = df.drop(['Z_CostContact', 'Z_Revenue', 'Dt_Customer'], axis=1)
```

```python
floats, objects = [], []
for col in df.columns:
    if df[col].dtype == object:
        objects.append(col)
    elif df[col].dtype == float:
        floats.append(col)

print(objects)
print(floats)
```

```python
for col in df.columns:
    if df[col].dtype == object:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
```

```python
scalar = StandardScaler()
data = scalar.fit_transform(df)
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2, random_state=0)
tsne_data = model.fit_transform(df)
plt.figure(figsize=(10,5))
plt.scatter(tsne_data[:, 0], tsne_data[:, 1])
plt.show()
```

```python
error = []
for n_clusters in range(1, 21):
    model = KMeans(init='k-means++',
                   n_clusters= n_clusters,
                   max_iter= 500,
                   random_state= 20)
    model.fit(df)
    error.append(model.inertia_)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=range(1,21), y=error)
sns.scatterplot(x=range(1,21), y=error)
plt.show()
```

```python
model = KMeans(init='k-means++',
                   n_clusters= n_clusters,
                   max_iter= 5,
                   random_state= 20)
segments = model.fit(df)
```

```python
plt.figure(figsize=(10,5))
df_tsne = pd.DataFrame({'x': tsne_data[:, 0], 'y': tsne_data[:, 1], 'segment': segments})
sns.scatterplot(x='x', y='y', hue='segment', data=df_tsne)
plt.show()
```
