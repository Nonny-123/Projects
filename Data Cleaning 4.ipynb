```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
df.head(10)
```

```python
df['Medal'] = df['Medal'].fillna("No Medal")
df['Height'] = df['Height'].fillna(df['Height'].median())
df['Weight'] = df['Weight'].fillna(df['Weight'].median())
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Height'] = df['Height'].astype(int)
df['Weight'] = df['Weight'].astype(int)
df['Age'] = df['Age'].astype(int)
```

```python
df.isnull().sum()
```

```python
df['Team'] = df['Team'].replace("Denmark/Sweden", "Denmark")
```

```python
df.info()
```

```python
df1.isnull().sum()
```

```python
df1.drop('notes', axis=1, inplace=True)
```

```python
df1.head(30)
```

```python
df1 = df1.dropna()
```

```python
df2 = pd.merge(df, df1, on='NOC', how='left')
```

```python
df3_Gold = df3[df3['Medal'] == 'Gold']
df3_Bronze = df3[df3['Medal'] == 'Bronze']
df3_Silver = df3[df3['Medal'] == 'Silver']
```

```python
team_medals = medals_df.groupby('Team')['Medal'].count()[:10]
```

```python
gender_medals = medals_df.groupby(['Sex'])['Medal'].count()
```

```python
df3_Gold.groupby('Team')['Medal'].count().sort_values(ascending=True)
```

```python
team_medals.sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 6))
df3_Gold.groupby('Team')['Medal'].count()[:10].sort_values(ascending=True).plot(marker='*', c='yellow')
df3_Bronze.groupby('Team')['Medal'].count()[:10].sort_values(ascending=True).plot(marker='o', c='brown')
df3_Silver.groupby('Team')['Medal'].count()[:10].sort_values(ascending=True).plot(marker='x', c='blue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
team_medals.sort_values(ascending=True).plot(kind='barh', color='skyblue')
plt.title('Top 10 Teams by Medal Count')
plt.xlabel('Number of Medals')
plt.ylabel('Team')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=gender_medals, y=gender_medals)
plt.title('Medals by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.show()
```
